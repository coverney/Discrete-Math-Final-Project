```python
import networkx as nx
import pandas as pd
import numpy as np
import time
import pickle
import matplotlib.pyplot as plt
```

Import the dictionary of ASIN IDs as keys to product titles as values:

```python
file_name = "ID_map-11-18"
picklefile = open(file_name,'rb')
ID_map = pickle.load(picklefile)

# Function for getting the name based on an ASIN ID:
def get_product_name(ASIN):
    if ASIN in ID_map:
        return ID_map[ASIN]
    else:
        print("There was an error with %s", ASIN)
```

Import adjacency matrix saved as copurchases pickle file

```python
# open the file for writing
file_name = "copurchases-11-20"
picklefile = open(file_name,'rb')
copurchases = pickle.load(picklefile)
```

Take the first X nodes and remove all other connections and nodes 

```python
num_nodes = 20000

sub_copurchases = {}
nodes = set()
for key in copurchases:
    if len(nodes) == num_nodes:
        break
    nodes.add(key)
    sub_copurchases[key] = []
    for v in copurchases[key]:
        if len(nodes) == num_nodes:
            break
        nodes.add(v)
        sub_copurchases[key].append(v)
```

Make sure everything in `nodes` is a key in `sub_copurchases`

```python
for node in nodes:
    if node not in sub_copurchases:
        sub_copurchases[node] = []
        try:
            for v in copurchases[node]:
                if v in nodes:
                    sub_copurchases[node].append(v)
        except:
            continue
```

Visualize co-purchases adjacency matrix

20,000 Nodes

```python
G_copurchases2 = nx.DiGraph()
G_copurchases2.add_nodes_from(sub_copurchases.keys())
for k, v in sub_copurchases.items():
    for t in v:
        if not G_copurchases2.has_edge(k, t) and not G_copurchases2.has_edge(t, k):
            G_copurchases2.add_edges_from(([(k, t)]))
print(nx.info(G_copurchases2))
```

```python
start = time.time()
pos = nx.spring_layout(G_copurchases2)
node_color = [2000.0 * G_copurchases2.degree(v) for v in G_copurchases2]
plt.figure(figsize=(20,20))
_ = nx.draw_networkx(G_copurchases2, pos=pos, with_labels=False,
                 node_color=node_color, node_size=100)
_ = plt.axis('off')
end = time.time()
print('time:', end - start)
```
